```python
import pandas as pd
import numpy as np
from argparse import Namespace
from sklearn.metrics import classification_report
from transformers import pipeline
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")

model = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC")
```

```python
from tqdm import tqdm
import torch
from transformers import pipeline
from concurrent.futures import ThreadPoolExecutor
device = 0 if torch.cuda.is_available() else -1
device
```

```python
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer, device = device)
```

```python
df = pd.read_csv('mail.csv')  # 데이터 수정
df.dropna(inplace=True)
df = df.drop(df.index[10:])
df
```

```python
# 몬테 카를로 법칙을 이용한 weight 뽑기 함수
def weight_select(n_iterations, weight_range, evaluate_model):
    best_weight = None
    best_score = float('-inf')
    
    for _ in range(n_iterations):
        weight = np.random.uniform(weight_range[0], weight_range[1])
        print(weight)
        score = evaluate_model(weight)
        
        if score > best_score:
            best_score = score
            best_weight = weight
    
    return best_weight, best_score
```

```python
# 평가하기 위한 모델 돌리고 베스트 값 도출
def evaluate_model(weight):
    sent_score_ls = []
    pst_ls = []
    neg_ls = []
    neut_ls = []
    flag = 0
    whyerror = []
    
    for i in tqdm(range(len(df)), desc="Processing articles"): ## 여기 모델 수정해야합니다.
        try:
            content = df['content'][i].split('. ')
            content = [e for e in content if len(e) <= 512]

            sent_ls = []
            sentiment = classifier(content)

            for j in range(len(sentiment)):
                sent_ls.append(sentiment[j]['label'])

            title_sentiment = classifier(df['title'][i])
            sent_ls.append(title_sentiment[0]['label'])
            neutral = sent_ls.count('neutral')
            positive = sent_ls.count('positive')
            negative = sent_ls.count('negative')
            neut_ls.append(neutral)
            pst_ls.append(positive)
            neg_ls.append(negative)
            sent_score = weight * positive - negative #sent_Score이 높은 거 할ㄹ거
            sent_score_ls.append(sent_score)
            
        except KeyError:
            pst_ls.append(0)
            neut_ls.append(0)
            neg_ls.append(0)
            sent_score_ls.append(0)
            whyerror.append(content)
            flag += 1

    # 여기서는 단순히 총 점수를 반환하지만, 실제로는 적절한 성능 평가 기준을 사용해야 합니다.
    return sum(sent_score_ls)
```

```python
# 몬테카를로 방법으로 최적의 가중치 선택
n_iterations = 10
weight_range = (0, 1)
best_weight, best_score = weight_select(n_iterations, weight_range, evaluate_model)

print(f"Best weight: {best_weight}, Best score: {best_score}")
```

```python
# 최종 모델 선정(best_weight 이용해서 여기 모델 수정하기)
from tqdm import tqdm

sent_score_ls = []
pst_ls = []; flag=0
neg_ls = []; neut_ls=[]; whyerror=[];

# tqdm을 사용하여 진행 상황 표시
for i in tqdm(range(len(df)), desc="Processing articles"):
    try:
        content = df['content'][i].split('. ')
        content = [e for e in content if len(e) <= 512]

        sent_ls = []
        sentiment = classifier(content)

        for j in range(len(sentiment)):
            sent_ls.append(sentiment[j]['label'])

        title_sentiment = classifier(df['title'][i])
        sent_ls.append(title_sentiment[0]['label'])
        neutral = sent_ls.count('neutral')
        positive = sent_ls.count('positive')
        negative = sent_ls.count('negative')
        neut_ls.append(neutral)
        pst_ls.append(positive)
        neg_ls.append(negative)
        sent_score = best_weight * positive - negative #위에서 best_weight 나온 걸로 sent_score 뽑아욤
        sent_score_ls.append(sent_score)
        
    except KeyError:
        pst_ls.append(0)
        neut_ls.append(0)
        neg_ls.append(0)
        sent_score_ls.append(0)
        whyerror.append(content)
        flag +=1
```

```python
print(pst_ls,neg_ls,neut_ls,sent_score_ls,flag,sep='\n')
```

```python
df['senti'] = sent_score_ls
df['positive'] = pst_ls
df['negative'] = neg_ls
df
```

```python
import pandas as pd
from tqdm import tqdm
from transformers import pipeline
from multiprocessing import Pool, cpu_count

# 감정 분석 모델 로드
classifier = pipeline('sentiment-analysis')

# 감정 분석 함수 정의
def analyze_sentiment(index):
    try:
        content = df['content'][index].split('. ')
        content = [e for e in content if len(e) <= 512]
        sent_ls = []

        # 배치 처리
        sentiments = classifier(content)
        sent_ls.extend([s['label'] for s in sentiments])

        title_sentiment = classifier(df['title'][index])
        sent_ls.append(title_sentiment[0]['label'])

        positive = sent_ls.count('positive')
        negative = sent_ls.count('negative')
        sent_score = positive - negative
        return sent_score
    except KeyError:
        return None


if __name__ == '__main__':
    indices = list(range(len(df)))
    with Pool(cpu_count()) as p:
        sent_score_ls = list(tqdm(p.imap(analyze_sentiment, indices), total=10, desc="Processing articles"))
    sent_score_ls = [score for score in sent_score_ls if score is not None]
    print(sent_score_ls)
```
