```python
import matplotlib.pyplot as plt

# Plotting
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.show()
```

```python
import pandas as pd
df_return = pd.read_excel("./csv/month_return.xlsx", index_col=0, parse_dates=True)/100
df_return
```

```python
mean_returns = df_return.mean()

# Calculate the covariance matrix
cov_matrix = df_return.cov()

# Display the results
print("Mean Returns:\n", mean_returns)
print("\nCovariance Matrix:\n", cov_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting mean returns as a bar chart
plt.figure(figsize=(10, 5))
mean_returns.plot(kind='bar', color='skyblue', edgecolor='black')
# plt.title('Mean Monthly Returns of ETFs')
# plt.xlabel('ETF Sectors')
# plt.ylabel('Mean Return')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()

# Plotting the covariance matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cov_matrix, annot=True, cmap='Blues', fmt='.4f', linewidths=0.5)
# plt.title('Covariance Matrix of ETF Returns')
plt.show()
```
